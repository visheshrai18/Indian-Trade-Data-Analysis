```python
import pandas as pd

# Load the datasets
import_data_df = pd.read_csv('/content/reverted_import_data.csv')
export_data_df = pd.read_csv('/content/reverted_export_data.csv')

# Descriptive statistics
print(import_data_df.describe())
print(export_data_df.describe())
```

```python
import matplotlib.pyplot as plt

# Original years list with '_x' suffix
years = ['2008-2009', '2009-2010_x', '2010-2011_x', '2011-2012_x', '2012-2013_x',
         '2013-2014_x', '2014-2015_x', '2015-2016_x', '2016-2017_x', '2017-2018_x',
         '2018-2019_x', '2019-2020_x', '2020-2021_x', '2021-2022_x', '2022-2023_x',
         '2023-2024_x',]

# Clean up the years list to remove '_x' suffix
cleaned_years = [year.replace('_x', '') for year in years]

# Plot total imports and exports by year
import_trend = import_data_df[years].sum()
export_trend = export_data_df[years].sum()

plt.figure(figsize=(12, 6))
plt.plot(cleaned_years, import_trend, label='Total Imports')
plt.plot(cleaned_years, export_trend, label='Total Exports')
plt.xticks(rotation=45)
plt.title('Yearly Trend of Imports and Exports')
plt.xlabel('Year')
plt.ylabel('Total Value (Rs Lac)')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

import_trend = import_data_df[years].sum()
export_trend = export_data_df[years].sum()

# Set the bar width
bar_width = 0.35

# Positions of the bars on the x-axis
r1 = np.arange(len(cleaned_years))
r2 = [x + bar_width for x in r1]

plt.figure(figsize=(12, 6))

# Create bar plots
plt.bar(r1, import_trend, color='blue', width=bar_width, label='Total Imports')
plt.bar(r2, export_trend, color='green', width=bar_width, label='Total Exports')

# Add labels
plt.xticks([r + bar_width/2 for r in range(len(cleaned_years))], cleaned_years, rotation=45)
plt.title('Yearly Trend of Imports and Exports')
plt.xlabel('Year')
plt.ylabel('Total Value (Rs Lac)')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd

# Assuming you have already loaded the import data into import_data_df

# Define the columns that represent the import values across the years
import_value_columns = ['2008-2009', '2009-2010_x', '2010-2011_x', '2011-2012_x',
                        '2012-2013_x', '2013-2014_x', '2014-2015_x', '2015-2016_x',
                        '2016-2017_x', '2017-2018_x', '2018-2019_x', '2019-2020_x',
                        '2020-2021_x', '2021-2022_x', '2022-2023_x', '2023-2024_x']

# Calculate the total import value for each country across all years
import_data_df['Total Import Value'] = import_data_df[import_value_columns].sum(axis=1)

# Calculate the overall market share for each country
total_import_sum = import_data_df['Total Import Value'].sum()
import_data_df['Overall Market Share (%)'] = (import_data_df['Total Import Value'] / total_import_sum) * 100

# Sort by market share
market_share_by_country = import_data_df[['Country', 'Total Import Value', 'Overall Market Share (%)']].sort_values(by='Overall Market Share (%)', ascending=False)

# Display the result
print(market_share_by_country.head(10))  # Display the top 10 countries by overall market share

# Define the columns that represent the export values across the years
export_value_columns = ['2008-2009', '2009-2010_x', '2010-2011_x', '2011-2012_x',
                        '2012-2013_x', '2013-2014_x', '2014-2015_x', '2015-2016_x',
                        '2016-2017_x', '2017-2018_x', '2018-2019_x', '2019-2020_x',
                        '2020-2021_x', '2021-2022_x', '2022-2023_x', '2023-2024_x']

# Calculate the total export value for each country across all years
export_data_df['Total Export Value'] = export_data_df[export_value_columns].sum(axis=1)

# Calculate the overall market share for each country
total_export_sum = export_data_df['Total Export Value'].sum()
export_data_df['Overall Market Share (%)'] = (export_data_df['Total Export Value'] / total_export_sum) * 100

# Sort by market share
market_share_by_export_country = export_data_df[['Country', 'Total Export Value', 'Overall Market Share (%)']].sort_values(by='Overall Market Share (%)', ascending=False)

# Display the result
print(market_share_by_export_country.head(10))  # Display the top 10 countries by overall market share
```

## Trade Intensity Index

```python
# Calculate Trade Intensity Index (Total Trade as a percentage of GDP)
gdp_data_df['Trade Intensity Index (%)'] = ((gdp_data_df['total_import'] + gdp_data_df['total_export']) / gdp_data_df['GDP (Million Dollars)']) * 100

# Display the result
print(gdp_data_df[['Year', 'Trade Intensity Index (%)']])

# Visualization
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Trade Intensity Index (%)', data=gdp_data_df, marker='o')
plt.title('Trade Intensity Index Over Time')
plt.ylabel('Trade Intensity Index (%)')
plt.xlabel('Year')
plt.show()
```

### 1. **Short Note on the Analysis**
   - **Trade Intensity Index Analysis**: The Trade Intensity Index measures the ratio of a country's total trade (imports + exports) to its GDP, expressed as a percentage. This analysis helps gauge the degree of an economy's openness and reliance on international trade relative to its overall economic output.

### 2. **Interpretation of the Result**
   - **Fluctuations**: The Trade Intensity Index for India has fluctuated between 19.64% in 2009 and 33.33% in 2022. After a dip in 2009, the index gradually increased, peaking in 2022. There was a notable decline from 2014 to 2017, followed by a sharp increase post-2020, with the index reaching its highest levels in 2022 and 2023.
   - **Recent Increase**: The significant rise from 2020 onwards could be attributed to changes in global trade dynamics, possibly influenced by India's increased participation in global markets or shifts in domestic economic policies.

### 3. **Opinion and Recommendations**
   - **Leverage Trade Opportunities**: The recent surge in the Trade Intensity Index suggests a growing reliance on international trade. India should capitalize on this by strengthening its trade agreements, improving trade infrastructure, and diversifying its trade partners to maintain and enhance this momentum.
   - **Manage Trade Dependency**: While increased trade intensity is beneficial, it also makes the economy more vulnerable to global market fluctuations. India should consider strategies to balance trade dependency with domestic economic growth to ensure sustainable long-term development.

There aren't fixed limits or universally agreed thresholds for what constitutes a "high" or "low" Trade Intensity Index (TII) because it can vary significantly depending on the country and its economic context. However, here's a general guideline:

- **Low TII (Below 20%)**: Typically indicates that the country has a relatively closed economy, with a focus on domestic production and less reliance on international trade.

- **Moderate TII (20% - 40%)**: Suggests a balanced economy where trade is important but not overwhelmingly dominant. This range is common for many developing and mid-sized economies.

- **High TII (Above 40%)**: Indicates that the country is highly integrated into the global economy, relying heavily on trade. This is often seen in small, open economies like Singapore or export-driven countries like Germany.

For large economies like India, a TII above 30% could be considered relatively high, reflecting significant engagement in international trade. However, the interpretation depends on the specific economic structure and goals of the country.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Determine the optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(trade_data_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()
```

## Multivariate Time Series Analysis

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt

# Load the data
file_path = '/content/converted_gdp_data.csv'
gdp_data_df = pd.read_csv(file_path)

# Select relevant economic indicators
data = gdp_data_df[['Year', 'GDP growth (annual%)', 'trade_balance', 'total_import', 'total_export']]

# Set 'Year' as the index
data.set_index('Year', inplace=True)

# Check for stationarity using the Augmented Dickey-Fuller test
def check_stationarity(timeseries):
    result = adfuller(timeseries)
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    for key, value in result[4].items():
        print('Critical Values:')
        print(f'   {key}, {value}')

# Apply the test on each series
for column in data.columns:
    print(f'\n{column}')
    check_stationarity(data[column])
    print('-----------------------')
```

```python
# Differentiating the data to make it stationary (if required)
data_diff = data.diff().dropna()

# Re-check for stationarity
for column in data_diff.columns:
    print(f'\n{column} (Differenced)')
    check_stationarity(data_diff[column])
    print('-----------------------')
```

## Vector Autoregression (VAR)

```python
from statsmodels.tsa.api import VAR

# Fit the VAR model
model = VAR(data_diff)
model_fit = model.fit(maxlags=4)  # You can change the lag value based on AIC/BIC criteria

# Print model summary
print(model_fit.summary())

# Forecast the next 5 periods
forecast = model_fit.forecast(data_diff.values[-model_fit.k_ar:], steps=5)
forecast_df = pd.DataFrame(forecast, index=np.arange(data.index[-1] + 1, data.index[-1] + 6), columns=data_diff.columns)

# Invert the differencing to obtain the forecast in the original scale
forecast_df = forecast_df.cumsum() + data.iloc[-1, :]

# Display the forecasted values
print(forecast_df)

# Visualization
forecast_df.plot(figsize=(12, 6))
plt.title('VAR Forecast of Economic Indicators')
plt.show()
```

## Multivariate LSTM

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

# Scale the data for LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Prepare the data for LSTM
def create_dataset(dataset, look_back=1):
    X, Y = [], []
    for i in range(len(dataset)-look_back):
        X.append(dataset[i:(i+look_back)])
        Y.append(dataset[i + look_back])
    return np.array(X), np.array(Y)

look_back = 3
X, Y = create_dataset(scaled_data, look_back)

# Reshape input to be [samples, time steps, features]
X = np.reshape(X, (X.shape[0], X.shape[1], X.shape[2]))

# Build the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(look_back, X.shape[2])))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(X.shape[2]))

model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X, Y, epochs=20, batch_size=1)

# Forecast the next 5 periods
predictions = model.predict(X[-5:])

# Invert scaling
predictions = scaler.inverse_transform(predictions)

# Create a DataFrame for the forecasted values
forecast_lstm_df = pd.DataFrame(predictions, index=np.arange(data.index[-1] + 1, data.index[-1] + 6), columns=data.columns)

# Display the forecasted values
print(forecast_lstm_df)

# Visualization
forecast_lstm_df.plot(figsize=(12, 6))
plt.title('LSTM Forecast of Economic Indicators')
plt.show()
```
